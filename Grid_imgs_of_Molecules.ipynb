```python
import os
from IPython.display import display

from rdkit import Chem
from rdkit.Chem import Draw
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
!ls "/content/drive/My Drive/Datasets_MolBio"
path_grid_sdf="/content/drive/My Drive/Datasets_MolBio/grid_mols.sdf"
```

```python
#Import molecules
if os.path.exists(path_grid_sdf):
  mols=Chem.SDMolSupplier(path_grid_sdf)
  mols=[mol for mol in mols if mol]
  print(len(mols))
else:
  print(f"SDF file not found: {path_grid_sdf}")
```

```python
#Draw molecules
img=Draw.MolsToGridImage(mols)
display(img)
```

```python
#Draw molecules parameters
img=Draw.MolsToGridImage(mols, subImgSize=(350,300), molsPerRow=2)
display(img)
```

```python
#Draw molecules drawing options
draw_opts=Draw.MolDrawOptions()
draw_opts.bondLineWidth=3
draw_opts.addAtomIndices = True
draw_opts.setBackgroundColour((1,1,0))
img=Draw.MolsToGridImage(mols, subImgSize=(350,300), molsPerRow=4, drawOptions=draw_opts)
display(img)
```

```python
#Draw molecules add Legends
titles=['Molecule A', 'Molecule B', 'Molecule C', 'Molecule D', 'Molecule E','Molecule F', 'Molecule G', 'Molecule H', 'Molecule I']
#img=Draw.MolsToGridImage(mols, subImgSize=(350,300), molsPerRow=4, legends=['Mol'+str(i) for i in range(len(mols))])
img=Draw.MolsToGridImage(mols, subImgSize=(350,300), molsPerRow=4, legends=titles)
display(img)
```

```python
#Draw molecules highlighting atoms
atom_hl_lists=[[0,1,2],[2,3,6],[],[4,5,6,7],[],[],[],[],[]]
img=Draw.MolsToGridImage(mols, subImgSize=(350,300), molsPerRow=4, highlightAtomLists=atom_hl_lists)
display(img)
```
