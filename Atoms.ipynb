```python
from rdkit import Chem
from rdkit.Chem import Draw
mol=Chem.MolFromSmiles('CCCNO')
print(mol)
Draw.MolToImage(mol)
```

```python
#Accessing Atoms
atoms=mol.GetAtoms()
#atoms- iterator object
print(atoms)
print(f" Number of atoms: {len(atoms)}")
num_atoms=mol.GetNumAtoms()
print(f" Number of atoms - GetNumAtoms(): {num_atoms}")

# Iterate over atoms
for atom in mol.GetAtoms():
    idx = atom.GetIdx()
    symbol = atom.GetSymbol()

    # Get valence information
    explicit_valence = atom.GetExplicitValence()  # Explicit valence (from bonds)
    implicit_valence = atom.GetImplicitValence()  # Implicit valence (from implicit hydrogens)
    total_valence = atom.GetTotalValence()       # Total valence (explicit + implicit)

    print(f"Atom index: {idx}, Symbol: {symbol}")
    print(f"  Explicit Valence: {explicit_valence}")
    print(f"  Implicit Valence: {implicit_valence}")
    print(f"  Total Valence: {total_valence}")
```

```python
#Accessing individual atoms
atom_1=mol.GetAtomWithIdx(1)
print(atom_1)
#for example
atom_1.SetFormalCharge(2)
Draw.MolToImage(mol)
```
