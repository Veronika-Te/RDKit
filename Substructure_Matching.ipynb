```python
from rdkit import Chem
from rdkit.Chem import Draw
mol=Chem.MolFromSmiles("C1CCCCC1NC2CCCCC2")
Draw.MolToImage(mol)
```

```python
def draw_mol_with_indices(mol):
  """Drawing the molecule with the indices."""
  opts=Draw.MolDrawOptions()
  opts.addAtomIndices =True
  return Draw.MolToImage(mol,options=opts, size=(400,400))

draw_mol_with_indices(mol)
```

```python
query_mol_1=Chem.MolFromSmiles('C1CCCCC1')
Draw.MolsToImage([query_mol_1])
```

```python
query_mol_2=Chem.MolFromSmiles('C1CCCCN1')
Draw.MolsToImage([query_mol_2])
```

Checking this molecule to see if it contains this substructure or not.

```python
#Substructure matching
mol.HasSubstructMatch(query_mol_1) #Output->bool: True
#mol.HasSubstructMatch(query_mol_2) #Output->bool: False
```

```python
match=mol.GetSubstructMatch(query_mol_1)
#Output->a tuple of the atom indices that match the query molecule.
print(match)
```

```python
matches=mol.GetSubstructMatches(query_mol_1)
print(matches) # ->
#This is a tuple of tuples of atom indexes.
#Each nested tuple corresponds to one match.
```

```python
from typing_extensions import Type
#SMARTS Pattern
smarts_pattern='C1CCCC[N,C]1' #[N,C] Nitrogen or Carbon
smarts_query_mol=Chem.MolFromSmarts(smarts_pattern)
#print(type(smarts_query_mol))
Draw.MolsToImage([smarts_query_mol])
```

```python
matches=mol.GetSubstructMatches(smarts_query_mol)
print(matches)
```
