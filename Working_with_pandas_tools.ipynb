```python
import os
from rdkit.Chem import PandasTools, Descriptors
import pandas as pd
```

```python

path='pandas_dataset.sdf'
if os.path.exists(path):
  df=PandasTools.LoadSDF('pandas_dataset.sdf', molColName='Mol')
  display(df)
else:
  print('File does not exist')

```

```python
print(df.shape)
```

```python
print(df.columns)
```

```python
mols=df['Mol']
print(mols)
```

```python
#Computing descriptors
df['Mol_Wt']=df['Mol'].apply(Descriptors.MolWt)
```

```python
display(df)
```

```python
#Sorting df
df.sort_values(by=['Mol_Wt'],ascending=False,inplace=True)
display(df)
```

Lead like compounds - 300 or less molecular weight

```python
#Filtering df
mask=df['Mol_Wt']<=300 #boolean values
#print(mask)
filtered_df=df[mask]
display(filtered_df)
```

```python
#Exporting as sdf file
PandasTools.WriteSDF(filtered_df,'filtered_df.sdf',molColName='Mol',properties=list(filtered_df.columns)) #✅Exported
```
